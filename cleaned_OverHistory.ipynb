```python
import pandas as pd 
import numpy as np

df = pd.read_csv('overHistory.csv')
df
```

```python

# Count occurrences of 'wd' and 'Nb' in each column
wd_nb_count = df.applymap(lambda x: str(x).count('Wd') + str(x).count('Nb')).sum()

print("Occurrences of 'wd' and 'Nb' in each column:")
print(wd_nb_count)
```

```python
# Filter rows to remove based on specified conditions
rows_to_remove = df[df.apply(lambda row: any(['Wd' in str(cell) or 'Nb' in str(cell) or any(char.isdigit() for char in str(cell)) and ('Wd' in str(cell) or 'Nb' in str(cell)) for cell in row]), axis=1)]

# Remove rows that meet the conditions
df_cleaned = df.drop(index=rows_to_remove.index)

print("DataFrame after removing rows with specified conditions:")
df_cleaned
```

```python
# List of columns to remove
columns_to_remove = ['ovBalls/6', 'ovBalls/7','ovBalls/8','ovBalls/9','ovBalls/10','ovBalls/11','ovBalls/12','ovBalls/13','bowlerId']

# Remove columns using drop()
df = df_cleaned.drop(columns=columns_to_remove)

df
```

```python
df['ovBalls/0'] = df['ovBalls/0'].str.replace('.', '0')
df['ovBalls/1'] = df['ovBalls/1'].str.replace('.', '0')
df['ovBalls/2'] = df['ovBalls/2'].str.replace('.', '0')
df['ovBalls/3'] = df['ovBalls/3'].str.replace('.', '0')
df['ovBalls/4'] = df['ovBalls/4'].str.replace('.', '0')
df['ovBalls/5'] = df['ovBalls/5'].str.replace('.', '0')
df
```
