```python
import pandas as pd 
import numpy as np
import re

df = pd.read_csv('matches.csv')
df
```

```python
columns_to_drop_range = df.columns[8:175]

# Drop columns within the specified range
df.drop(columns=columns_to_drop_range, inplace=True)
```

```python
df
```

```python
df.dtypes
```

```python
df['matchDate'] = pd.to_datetime(df['matchDate'],utc = True)
df['matchEndDate'] = pd.to_datetime(df['matchEndDate'],utc = True)

# Check the updated DataFrame and datatypes of the columns
print(df)
print(df.dtypes)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df = df.dropna(subset=['toss.elected', 'toss.winner'])
```

```python
df
```

```python
df.isnull().sum()
```

```python
# Forward fill missing values in a datetime column
df['matchEndDate'].fillna(method='ffill', inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
# Function to remove punctuation using regular expressions
def remove_punctuation_regex(text):
    return re.sub(r'[^\w\s]', '', text)

# Apply the function to the specified columns
df['toss.elected'] = df['toss.elected'].apply(remove_punctuation_regex)
df['toss.winner'] = df['toss.winner'].apply(remove_punctuation_regex)

print(df)
```

```python
df
```

```python
df['toss.winner'].value_counts()
```

```python
df['toss.winner'] = df['toss.winner'].replace('United States', 'USA')
```

```python
df['toss.winner'] = df['toss.winner'].replace('United States of America', 'USA')
```

```python
df['toss.winner'] = df['toss.winner'].replace('United Arab Emirates', 'UAE')
```

```python
df['toss.winner'] = df['toss.winner'].replace('East Africa', 'South Africa')
```

```python
df['toss.winner'] = df['toss.winner'].replace('Windies', 'West Indies')
```

```python
 df['toss.winner'] = df['toss.winner'].replace('West Indies', 'WestIndies')
```
